```python
import pandas as pd
```

```python
pci_1 = pd.read_csv("pci_al-fl.csv")
pci_1.set_index("DATE", inplace = True)
pci_1.head()
```

```python
pci_2 = pd.read_csv("pci_ga-me.csv")
pci_2.set_index("DATE", inplace = True)
pci_2.head()
```

```python
pci_3 = pd.read_csv("pci_md-nh.csv")
pci_3.set_index("DATE", inplace = True)
pci_3.head()
```

```python
pci_4 = pd.read_csv("pci_nj-ri.csv")
pci_4.set_index("DATE", inplace = True)
pci_4.head()
```

```python
pci_5 = pd.read_csv("pci_sc-wy.csv")
pci_5.set_index("DATE", inplace = True)
pci_5.head()
```

```python
pci_by_state_cleaned = pd.concat([pci_1, pci_2, pci_3, pci_4, pci_5], axis=1).head()
pci_by_state_cleaned.head()
```

```python
pci_by_state_cleaned = pci_by_state_cleaned.transpose()
print(pci_by_state_cleaned)
```

```python
pci_by_state_cleaned.dtypes
```

```python
pci_by_state_cleaned = pci_by_state_cleaned.rename(columns = {'2016-01-01': '2016', '2017-01-01': '2017', '2018-01-01': '2018', '2019-01-01': '2019'})
print (pci_by_state_cleaned)
```

```python
pci_by_state_cleaned.to_csv(r'pci_by_state_cleaned.csv')
```
